# Problem 3: Matrix Multiplication, Image Processing, and Photo Filters

```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt
```

```python
# Load an Image as a Matrix
img = imageio.imread("mount.jpg")
print("Image Dimensions:", img.shape)
```

```python
# Extracting Image Information
height, width, channels = img.shape
R = img[:, :, 0]  # Red channel
G = img[:, :, 1]  # Green channel
B = img[:, :, 2]  # Blue channel

plt.imshow(R, cmap='Reds')
plt.title("Red Channel")
plt.show()

plt.imshow(G, cmap='Greens')
plt.title("Green Channel")
plt.show()

plt.imshow(B, cmap='Blues')
plt.title("Blue Channel")
plt.show()
```

```python
# Convert Image to Grayscale
grayscale_img = (0.2989 * R + 0.5870 * G + 0.1140 * B).astype(np.uint8)
plt.imshow(grayscale_img, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```

```python
# Apply Sepia Filter
sepia_filter = np.array([[0.393, 0.769, 0.189],
                         [0.349, 0.686, 0.168],
                         [0.272, 0.534, 0.131]])
sepia_img = np.dot(img[..., :3], sepia_filter.T).clip(0, 255).astype(np.uint8)
plt.imshow(sepia_img)
plt.title("Sepia Image")
plt.show()
```

```python
# Extract Red Channel Only
red_only_img = img.copy()
red_only_img[:, :, 1:] = 0
plt.imshow(red_only_img)
plt.title("Red Channel Only")
plt.show()
```

```python
# Swap Red and Blue Channels
swapped_img = img.copy()
swapped_img[:, :, [0, 2]] = swapped_img[:, :, [2, 0]]
plt.imshow(swapped_img)
plt.title("Swapped Red and Blue")
plt.show()
```

```python
# Remove Green Channel
green_removed_img = img.copy()
green_removed_img[:, :, 1] = 0
plt.imshow(green_removed_img)
plt.title("Green Removed")
plt.show()
```

```python
# Color Inversion
inverted_img = 255 - img
plt.imshow(inverted_img)
plt.title("Inverted Image")
plt.show()
```

```python
# Modify Saturation
saturation_filter = np.array([[1.2, 0, 0],
                              [0, 0.8, 0],
                              [0, 0, 1.2]])
saturation_img = np.dot(img[..., :3], saturation_filter.T).clip(0, 255).astype(np.uint8)
plt.imshow(saturation_img)
plt.title("Modified Saturation")
plt.show()
```

```python
# Custom Color Adjustments
user_filter = np.array([[1.1, 0.1, 0.1],
                        [0.1, 0.9, 0.1],
                        [0.1, 0.1, 1.1]])
user_adjusted_img = np.dot(img[..., :3], user_filter.T).clip(0, 255).astype(np.uint8)
plt.imshow(user_adjusted_img)
plt.title("User Adjusted Image")
plt.show()
```

```python
# Matrix Inversion to Restore Original Image
inverse_filter = np.linalg.inv(user_filter)
restored_img = np.dot(user_adjusted_img[..., :3], inverse_filter.T).clip(0, 255).astype(np.uint8)
plt.imshow(restored_img)
plt.title("Restored Image")
plt.show()
```

```python
# Apply Gamma Correction
gamma = 2.2
gamma_corrected_img = (255 * ((img / 255) ** (1/gamma))).clip(0, 255).astype(np.uint8)
plt.imshow(gamma_corrected_img)
plt.title("Gamma Corrected Image")
plt.show()
```
