# Project 2: Matrix operations and image manipulation

```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt
```

```python
# Load the image
img = imageio.imread("einstein.jpg")
```

```python
# Check the dimensions of the image
print(
    "Image Dimensions:", img.shape
)
```

```python
# Check type
print("Data type of image array:", img.dtype)

if np.issubdtype(img.dtype, np.integer):
    print("The array elements are integers.")
else:
    print("The array elements are not integers.")
```

```python
# Convert to Grayscale
if len(img.shape) == 3:
    grayscale_img = np.mean(img, axis=2).astype(np.uint8)
else:
    grayscale_img = img
```

```python
# Lightening & Darkening the Image
lightened = np.clip(grayscale_img + 50, 0, 255)
darkened = np.clip(grayscale_img - 50, 0, 255)
```

```python
# Adjust Contrast
high_contrast = np.clip((grayscale_img - 128) * 1.5 + 128, 0, 255)
```

```python
# Crop the Image
cropped = grayscale_img[200:600, 200:500]
```

```python
# Flip and Rotate Image
flipped_img = np.flip(grayscale_img, axis=1)  # Flip horizontally
rotated_img = np.rot90(grayscale_img)  # Rotate 90 degrees
```

```python
# Create a list of images and their titles
images = [
    (img, "Original Image"),
    (grayscale_img, "Grayscale Image"),
    (lightened, "Lightened Image"),
    (darkened, "Darkened Image"),
    (high_contrast, "High Contrast Image"),
    (cropped, "Cropped Image"),
    (flipped_img, "Flipped Image"),
    (rotated_img, "Rotated Image"),
]

# Create a figure to display all images
fig, axes = plt.subplots(2, 4, figsize=(20, 10))  # 2 rows, 4 columns
axes = axes.ravel()

# Use a for loop to display each image
for i, (image, title) in enumerate(images):
    if len(image.shape) == 3:  # Check if the image is RGB
        axes[i].imshow(image)
    else:  # Grayscale image
        axes[i].imshow(image, cmap="gray")
    axes[i].set_title(title)
    axes[i].axis("off")  # Hide axes

# Adjust layout and show the figure
plt.tight_layout()
plt.show()
```
