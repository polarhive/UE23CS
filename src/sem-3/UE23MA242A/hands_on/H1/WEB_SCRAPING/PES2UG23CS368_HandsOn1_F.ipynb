```python
from bs4 import BeautifulSoup as bs
import requests

#Target URL
url = 'https://www.amazon.in/Apple-iPhone-128GB-Deep-Purple/product-reviews/B0BDJH6GL8'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}
response = requests.get(url, headers=headers)

# Ensure the page was successfully fetched
if response.status_code == 200:
    print("Page fetched successfully")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
page=requests.get(url)
print(page)
```

```python
#next we will create a BeautifulSoup object called soup
soup=bs(page.content,'html.parser')

#print(soup)  #raw html data
print(soup.prettify())
```

```python
name=soup.find_all('span',class_='a-profile-name')
print(name)
```

```python
cust_name=[]

# .get_text() method is used to extract the textual content from a BeautifulSoup object.
for i in range(0,len(name)):
    cust_name.append(name[i].get_text())
cust_name
```

```python
title=soup.find_all('a', class_="review-title")
title
```

```python
review_title=[]
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title=[i.lstrip('5.0 out of 5 stars\n') for i in review_title]
review_title=[i.lstrip('4.0 out of 5 stars\n') for i in review_title]
review_title=[i.lstrip('3.0 out of 5 stars\n') for i in review_title]
review_title=[i.lstrip('1.0 out of 5 stars\n') for i in review_title]

review_title1=[i.rstrip('\n')  for i in review_title]
review_title1
```

```python
rating=soup.find_all('i',class_='review-rating')
cust_rating=[]
for i in range(0,len(rating)):
    cust_rating.append(rating[i].get_text())
cust_rating
```

```python
content=soup.find_all("span",class_="review-text-content")
cust_content=[]
for i in range(0,len(content)):
    cust_content.append(content[i].get_text())
cust_content1=[i.lstrip('\n') for i in cust_content]
cust_content2=[i.rstrip('\n')  for i in cust_content1]
cust_content2
```

```python
import pandas as pd

# Ensure all lists have the same length before creating the DataFrame
min_length = min(len(cust_name), len(review_title1), len(cust_content2))

df = pd.DataFrame({
    'cust_name': cust_name[:min_length],
    'review_title1': review_title1[:min_length],

    'cust_content': cust_content2[:min_length]
})

df
```

```python
df.to_csv(r'amazon.csv')
```
